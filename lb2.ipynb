```python
import decimal as dec
a = dec.Decimal('1')
b = dec.Decimal('7')
print(a+b)

dec.getcontext().prec = 5
print(a/b)
print((a/b).quantize(dec.Decimal('0.0000')))

import fractions as frac

c = frac.Fraction(1, 3)
d = frac.Fraction(4, 3)
e = frac.Fraction(0.1)

print(c + d)
print(e)

print(c.denominator)
print(d.numerator)
print(frac.Fraction(4.1212121212).limit_denominator(10))
```

```python
S = 'spam' 
S = S[:1] + 'l' + S[2:]
print(S)

S = S[0]+ 'l' + S[2] + S[3]
print(S)

S = 'spam'
s  =  list(S)
s[1] = "l"
S = ''.join(s)
print(S)
"""
Со списокм такая же логика, но при присваивании по индексам, понятно, что перереводить в список не надо. С кортежами разрезание и конкатенация также работают, однако напрямую изменить элемент в силу немутабельности нельзя( как и в строке - пришлось переводить в список).
"""
```
